# Q1: Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?

Ex:

stutter("incredible") ➞ "in... in... incredible?"

stutter("enthusiastic") ➞ "en... en... enthusiastic?"

stutter("outstanding") ➞ "ou... ou... outstanding?"

Hint :- Assume all input is in lower case and at least two characters long.

```python
def sutter():
    stuttering_word = str(input("Enter a word: ")).lower()
    two_letters = stuttering_word[:2]+str("...")
    print("\nSomeone with a stammer would say",stuttering_word,"as:",two_letters,two_letters,"...",stuttering_word)
sutter()
```

# Q2: Question 2.Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.

Ex:

radians_to_degrees(1) ➞ 57.3

radians_to_degrees(20) ➞ 1145.9

radians_to_degrees(50) ➞ 2864.8

```python
def radians_to_degrees():
    radians = float(input("Enter angles in radians: "))
    radians_to_degrees_conversion = str(round((radians*57.2958),1))+str("°")
    print(radians,"in degrees is:",radians_to_degrees_conversion)
radians_to_degrees()
```

# Q3: In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.

Given a non-negative integer num, implement a function that returns True if num is a Curzon number, or False otherwise.

Ex:

is_curzon(5) ➞ True since 2 * 5 + 1 = 33 2 5 + 1 = 11 33 is a multiple of 11

is_curzon(10) ➞ False since 2 * 10 + 1 = 1025 2 10 + 1 = 21 1025 is not a multiple of 21

is_curzon(14) ➞ True since 2 * 14 + 1 = 16385 2 14 + 1 = 29 16385 is a multiple of 29

```python
def is_curzon():
    curzon_number = int(input("Enter a whole number: "))
    numerator = (2*curzon_number)+1
    denominator = (2**curzon_number)+1
    if denominator % numerator == 0: print("True")
    else: print("False")
is_curzon()
```

# Q5: Create a function that returns a base-2 (binary) representation of a base-10 (decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10) 010101001(2) = 1 + 8 + 32 + 128.

Going from right to left, the value of the most right bit is 1, now from that every bit to the left will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1). Ex: binary(1) ➞ "1"

1*1 = 1

binary(5) ➞ "101"

11 + 14 = 5

binary(10) ➞ "1010"

12 + 18 = 10

```python
def binary():
    number_base = int(input("Enter a number:"))
    b_string  = bin(number_base)     
    b_string =  b_string[2:] 
    print(str(number_base)+"'s base(2) conversion is:",b_string)
    integer_conversion = 0  
    for idx,val in enumerate(b_string[::-1]):   
        if ((int(val) & 1 ) == 1):              
            integer_conversion = 1*(1<<idx)+integer_conversion 
    print("Converting "+str(b_string),"back to it's base(10) form:",integer_conversion)  
binary()
```
