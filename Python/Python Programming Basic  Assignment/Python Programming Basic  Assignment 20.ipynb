# Q2: Given a list of numbers, create a function which returns the list but with each element's index in the list added to itself. This means you add 0 to the number at index 0, add 1 to the number at index 1, etc...

Ex:

i. add_indexes([0, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4]

ii. add_indexes([1, 2, 3, 4, 5]) ➞ [1, 3, 5, 7, 9]

iii. add_indexes([5, 4, 3, 2, 1]) ➞ [5, 5, 5, 5, 5]

```python
add_indexes = []
upper = int(input("Enter the range for the list: "))
l1 =[int(input("Enter a number: ")) for x in range(upper)]
for value in l1:
    c = value + l1.index(value)
    add_indexes.append(c)
print(add_indexes)
```

This means that the first triangle has just one dot, the second one has three dots, the third one has 6 dots and so on. Write a function that gives the number of dots with its corresponding triangle number of the sequence.

Ex:

i. triangle(1) ➞ 1

ii. triangle(6) ➞ 21

iii. triangle(215) ➞ 23220

```python
def triangle():
    dot = int(input("Enter the total number of dots in the triangle: "))
    print("The triangular number for",dot,"number of triangles is - ",((dot*(dot+1)/2)))
triangle()
```
